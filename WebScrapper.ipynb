```python
def getRoles(link):
  hero_link = link
  dota_page = requests.get(hero_link)
  dota_list = BeautifulSoup(dota_page.content, "html.parser")
  dota_list2 = dota_list.find(id="hero")
  hero_name = dota_list2.find("div", class_="role-box")

  p = list(hero_name.children)[1]
  res = p.get_text()

  output = [] ##0: Faction, 1: MainAttr, 2: Attack 
  
  faction1 = "Dire"
  Abil1 = "Strength"
  Abil2 = "Agility"
  Attack = "Range"

  if faction1 in res: 
    output.append(faction1)
  else: 
    output.append("Radiant")

  if Abil1 in res: 
    output.append(Abil1)
  elif Abil2 in res: 
    output.append(Abil2)
  else: 
    output.append("Intelligence")

  if Attack in res: 
    output.append(Attack)
  else: 
    output.append("Melee")
  
  return output

def getName(link): 
  hero_link = link
  dota_page = requests.get(hero_link)
  dota_list = BeautifulSoup(dota_page.content, "html.parser")
  dota_list2 = dota_list.find(id="hero")
  hero_name_src = dota_list2.find("h1", class_="c")
  name = list(hero_name_src.children)[3]
  return name

def getStats_util(data):
  res = data.get_text().split(" ")
  output = []
  for i in res: 
    s = i.strip("\n")
    r = s.replace("\n", " ")
    output.append(r)
  return output 


def getStats(link):
  hero_link = link
  dota_page = requests.get(hero_link)
  dota_list = BeautifulSoup(dota_page.content, "html.parser")
  dota_list2 = dota_list.find(id="hero")
  stats_data1 = dota_list2.find_all("tr", class_="bg1")
  stats_data2 = dota_list2.find_all("tr", class_="bg2")
  output = []
  for i in stats_data1: 
    temp = getStats_util(i)
    output.append(temp)

  for i in stats_data2: 
    temp = getStats_util(i)
    output.append(temp)
  return output

def getLvls(link): 
  hero_link = link
  dota_page = requests.get(hero_link)
  dota_list = BeautifulSoup(dota_page.content, "html.parser")
  dota_list2 = dota_list.find(id="hero")
  levels = dota_list2.find_all("div", class_="hero-view-talents")

  output = []

  for i in levels: 
    temp = i.get_text()
    temp2 = temp.split(" ")
    res = []
    for i in temp2: 
      s = i.strip("\n")
      r = s.replace("\n", " ")
      res.append(r)
    output.append(res)

  return output

def getLore(link): 
  hero_link = link
  dota_page = requests.get(hero_link)
  dota_list = BeautifulSoup(dota_page.content, "html.parser")
  dota_list2 = dota_list.find(id="hero")
  lore = dota_list2.find("div", class_="bio-wrap self-clear bio-height")
  n = lore.get_text().strip("\n")
  output = n.replace("\n", " ")
  return output

def getAbil_Util_Link(link):
  
  herolink = link
  dota_page = requests.get(link)
  dota_list = BeautifulSoup(dota_page.content, "html.parser")
  dota_list2 = dota_list.find(id="hero")
  abil_data = dota_list2.find("div", class_="skills-box self-clear mt10")
  abil_link = abil_data.find_all("a", href=True)
  list(abil_link)

  ability = []

  for link in abil_link: 
    size = len(link['href'])
    ability.append(link['href'][14:size])

  return ability

def getAbil_Util_Data(link):
  herolink = link
  dota_page = requests.get(link)
  dota_list = BeautifulSoup(dota_page.content, "html.parser")

  abil_mana = dota_list.find_all("span", class_="hiliteG")
  mana_cd = list(abil_mana)
  output = []
  for i in mana_cd:
    output.append(i.get_text())
  abilMana = output

  abil_name = dota_list.find("h2", class_="hiliteW")
  abilName = abil_name.get_text()

  abil_data = dota_list.find("div", class_="mt10")
  abilData = abil_data.get_text().split("\n")

  res = [ abilName, abilData, abilMana]
  return res

def getAbil(link): 
  link2 = "https://www.dotafire.com/dota-2/skill/"
  abil_links = getAbil_Util_Link(link)
  output = []
  for i in abil_links: 
     res = getAbil_Util_Data(link2+i)
     output.append(res)

  return output

  x = x + "0"

def getAbilImg():
  switch = False
  starting_hero = ""
  for z in range(len(hero_web)):
    if switch == False: 
      if hero_web[z+1] == starting_hero: 
        switch = True
      continue
    hero_name_now = hero_web[z]
    url1 = 'https://dota2.fandom.com/wiki/' + hero_name_now
    dota_page = requests.get(url1)
    dota_list = BeautifulSoup(dota_page.content, "html.parser")
    dota_list2 = dota_list.find("div", class_="mw-parser-output")
    images2 = dota_list2.find_all("img")
    output = []
    x = "0"
    for i in images2: 
      try:
        if i["alt"] == "Guide icon.png": 
          break
        if i["width"] == "128":
          if i["height"] == "128":
            output.append(i)
      except KeyError:
        continue
        
    list_images = []

    for i in output: 
      list_images.append(i["data-src"])


    for i in list_images: 
      link = i
      img = Image.open(requests.get(link, stream=True).raw)
      path = "/content/drive/MyDrive/Dota/Hero_img/" + hero_drive[z] + "/" + hero_name_now + x + ".png"
      img.save(path)
      x = x + "0"
    print(z)
  
  def getHeroData_and_Write():
    dota = "https://www.dotafire.com/dota-2/hero/"
    results = []
    for i in hero_links_names:
      hero_link = dota + i

      lore = getLore(hero_link)

      roles = getRoles(hero_link)

      name = getName(hero_link)

      stats = getStats(hero_link)

      abilities = getAbil(hero_link)

      levels = getLvls(hero_link)

      hero = {
      "name": name,
      "roles": roles,
      "Lore": lore, 
      "abilities": abilities,
      "lvl": levels,
      "stats": stats 
      };

      results.append(hero) 

    json_object = json.dumps(results, indent = 4)

    with open("/content/drive/MyDrive/Dota/hero_data.json", "w") as outfile:
          outfile.write(json_object)
  
```
